```python
# Decision Tree Example: Weather vs Activity

# 1️⃣ Import libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

# 2️⃣ Sample dataset
data = {
    'Weather': ['Sunny', 'Sunny', 'Rainy', 'Rainy', 'Sunny', 'Rainy'],
    'Activity': ['Go Out', 'Go Out', 'Stay In', 'Stay In', 'Go Out', 'Stay In']
}

df = pd.DataFrame(data)

# 3️⃣ Encode categorical variables
df_encoded = pd.get_dummies(df, drop_first=True)
# 'Weather_Rainy' and 'Activity_Stay In'

X = df_encoded.drop('Activity_Stay In', axis=1)
y = df_encoded['Activity_Stay In']

# 4️⃣ Train Decision Tree
model = DecisionTreeClassifier(criterion='entropy', random_state=42)
model.fit(X, y)

# 5️⃣ Plot the Decision Tree
plt.figure(figsize=(6,4))
plot_tree(
    model,
    feature_names=X.columns,
    class_names=['Go Out','Stay In'],
    filled=True
)
plt.title("Decision Tree for Weather vs Activity")
plt.show()

# 6️⃣ Optional: Check accuracy on training data
accuracy = model.score(X, y)
print(f"\nAccuracy on training data: {accuracy*100:.2f}%")
```
